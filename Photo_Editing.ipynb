```python
##### %matplotlib inline
import PIL
import matplotlib.pyplot as plt 
import os.path
import numpy as np
import random
import time
from IPython.display import clear_output


def edit():
    image = raw_input("Please input the name of the file you'ld like to edit.\n Example : Test.jpg\n")
    time.sleep(3)
    clear_output()
    
    directory = os.getcwd()
    filename = os.path.join(directory, str(image))
    img = plt.imread(filename)
    img.setflags(write=1)
    
    edit = int(raw_input("How would you like to edit your image?\n If you would like a color change, press 1.\n If you want The Rock to be in the picture, press 2.\n If you would like to lower the quality to insane amounts, press 3.\n If you would like it to be black and white, press 4.\n Lastly, if you would like to Have the Glarner Knight logo in the picture, press 5.\n"))
    time.sleep(3)
    clear_output()
    
    #Color Change
    if edit == 1:
        color = raw_input("What color of the rainbow would you like?\n")
        time.sleep(3)
        clear_output()
        
        # Change to Red
        if color.lower() == 'red':
            height = len(img)
            width = len(img[0])
            for r in range(height):
                for c in range(width):54
                    if sum(img[r][c])>500:
                        img[r][c]=[255,0,0]
        # Change to Green
        elif color.lower() == 'green':
            height = len(img)
            width = len(img[0])
            for r in range(height):
                for c in range(width):
                    if sum(img[r][c])>500:
                        img[r][c]=[0,255,0]
        # Change to Blue
        elif color.lower() == 'blue':
            height = len(img)
            width = len(img[0])
            for r in range(height):
                for c in range(width):
                    if sum(img[r][c])>500:
                        img[r][c]=[0,0,255]
        # Change to Yellow
        elif color.lower() == 'yellow':
            height = len(img)
            width = len(img[0])
            for r in range(height):
                for c in range(width):
                    if sum(img[r][c])>500:
                        img[r][c]=[255,255,51]
        # Change to Violet
        elif color.lower() == 'violet':
            height = len(img)
            width = len(img[0])
            for r in range(height):
                for c in range(width):
                    if sum(img[r][c])>500:
                        img[r][c]=[153,0,153]
        # Change to Orange
        elif color.lower() == 'orange':
            height = len(img)
            width = len(img[0])
            for r in range(height):
                for c in range(width):
                    if sum(img[r][c])>500:
                        img[r][c]=[255,153,0]
                        
    #Rocks
    elif edit == 2:
        directory = os.getcwd()
        img = PIL.Image.open(filename)
        File = os.path.join(directory, 'Rock.png')
        Image = PIL.Image.open(File)
        Small = Image.resize((75, 100))
        img.paste(Small, (300,0), mask=Small)
    
    #Pixelate
    elif edit == 3:
        directory = os.getcwd()
        filename = os.path.join(directory, str(image))
        Image = PIL.Image.open(filename)
        img = Image.resize((12, 12))
    
    #Black and White
    elif edit == 4:
        height = len(img)
        width = len(img[0])
        for r in range(height):
            for c in range(width):
                    img[r][c][1]=img[r][c][0]
                    img[r][c][2]=img[r][c][0]

    #Pride
    elif edit == 5:
        directory = os.getcwd()
        img = PIL.Image.open(filename)
        File = os.path.join(directory, 'NGSD.png')
        Image = PIL.Image.open(File)
        Small = Image.resize((50, 50))
        img.paste(Small, (int(random.randint(50,350)), int(random.randint(25,175))), mask=Small)
                    
    fig, ax = plt.subplots(1, 1)
    ax.imshow(img, interpolation='none')

    fig.show()
    
edit()
```
